## 3. 파생변수 만들기

```python
data = {
    'var1': [4, 3, 8],
    'var2': [2, 6, 1],
    'var3': [4, 2, 7],
    'var4': [1, 4, 5]
}
df = pd.DataFrame(data)
df
```

```python
df_new = df.copy()
df['var_sum'] = df['var1'] + df['var2'] + df['var3'] + df['var4']
# df['var_sum'] = sum([df[f'var{n}'] for n in range(1, 5)])
# df['var_sum'] = sum([df[column] for column in df.columns])
df['var_mean'] = df['var_sum'] / 4
df
```

```python
mpg['total'] = mpg['cty'] + mpg['hwy'] / 2
mpg['total'].plot.hist()
```

조건문 사용하기

```python
import numpy as np
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg['test'].value_counts().plot.bar(rot=0)
```

중첩 조건문

```python
mpg['test'] = np.where(mpg['total'] >= 40, 'A',
                np.where(mpg['total'] >= 20, 'B', 'C'))
print(mpg['test'].value_counts())
mpg['test'].value_counts().sort_index().plot.bar(rot=0)
```

포함하는지 확인

```python
print(mpg['category'].value_counts())

mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']),
                       'small', 'large')
mpg['size'].value_counts().sort_index(ascending=False).plot.bar(rot=0)
```
