## 2. 데이터 프레임 만들기

### 1. List (or Numpy array)

```python
import pandas as pd

data = {
    'name': ['Kim', 'Lee', 'Park', 'Choi'],
    'english': [90, 80, 60, 70],
    'math': [50, 60, 100, 20]
}

df = pd.DataFrame(data)

print(df)
```

### 2. Dictionary

```python
import pandas as pd

data = [
    ['Kim', 90, 50],
    ['Lee', 80, 60],
    ['Park', 60, 100],
    ['Choi', 70, 20]]

df = pd.DataFrame(
    data,
    columns=['name', 'english', 'math']
)

print(df)
```

### 3. Dictionary : index & column

```python
import pandas as pd

data = [
    {
        'name': 'Kim',
        'english': 90,
        'math': 50
    },
    {
        'name': 'Lee',
        'english': 80,
        'math': 60
    },
    {
        'name': 'Park',
        'english': 60,
        'math': 100
    },
    {
        'name': 'Choi',
        'english': 70,
        'math': 20
    }
]

df = pd.DataFrame(data)

print(df)
```
