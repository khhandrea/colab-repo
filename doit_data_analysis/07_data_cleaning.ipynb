## 2. 이상치 정제

```python
df = pd.DataFrame({
    'sex': [1, 2, 1, 3, 2, 1], # 1: male, 2: female
    'score': [5, 4, 3, 4, 2, 6]
})
df
```

이상치 확인하기

```python
df['sex'].value_counts().sort_index()
```

이상치 처리하기

```python
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df['sex'].value_counts().sort_index()
```

극단적인 값

```python
mpg = pd.read_csv('./files/mpg.csv')
mpg.head()
```

```python
import seaborn as sns
sns.boxplot(data = mpg, y = 'hwy')
```

```python
pct25 = mpg['hwy'].quantile(.25)
pct75 = mpg['hwy'].quantile(.75)
iqr = pct75 - pct25
quantile_min = pct25 - 1.5 * iqr
quantile_max = pct75 + 1.5 * iqr

mpg['hwy'] = np.where((mpg['hwy'] < quantile_min) | (mpg['hwy'] > quantile_max),
                      np.nan, mpg['hwy'])
mpg['hwy'].isna().value_counts()
```

```python
mpg.dropna(subset=['hwy'])\
    .groupby('drv')\
    .agg(mean_hwy=('hwy', 'mean'))
```
